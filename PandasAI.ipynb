```python
import pandas as pd
from pandasai import PandasAI
```

```python
# Instantiate a LLM
from pandasai.llm.openai import OpenAI
llm = OpenAI(api_token="Your_api_key")

pandas_ai = PandasAI(llm, conversational=False)
```

```python
df = pd.read_csv('insurance.csv')
df.head()
```

```python
pandas_ai.run(df, prompt='Plot a graph to count the number of smokers by sex.')
```

```python
pandas_ai.run(df, prompt='what is the average charges?')
```

```python
pandas_ai.run(df, prompt='Are there any null values?')
```

```python
pandas_ai.run(df, prompt="Plot a graph showing the distribution of smokers' ages and identify which region has the most smokers.'")
```
